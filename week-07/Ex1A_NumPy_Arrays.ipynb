## Looping Through Arrays

In a new code cell, create a variable named arrA as a two-dimensional array
containing (11, 12, 13) and (14, 15, 16) then print arrA.

```python
arrA = np.array([[11,12,13],[14,15,16]])
print(arrA)
```

In a new code cell, loop through the first dimension of arrA (i.e. for x in arrA) and
print each element multiplied by 2

```python
for x in arrA:
    print(x * 2)
```

In a new code cell, loop through the second dimension of arrA (i.e. for y in x) and
print each element multiplied by 2

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

In the first loop, it only went to the first dimension, so thats why you can still see the brackets. For the second loop, it went deeper to the second dimension and only displayed the values that was in the first dimension.

Add another code cell. Again loop through the second dimension of arrA (i.e. for y
in x), but this time, assign x*2 to a variable named loop_result. Add another line
of code to print loop_result, then add a comment to the cell to explain how/why
this result is different.

```python
for x in arrA:
    for y in x:
        loop_result = x*2
    
print(loop_result)

# The reason why the results are different, is because x is the accumulating the results but getting overwrited after each iteration
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
# I had a ValueError. It was an error because the arrays that are being stacked must have the same shape.
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
# ValueError! the input array dimensions axis doesn't match
```

```python
# the control. .stack()
control_stack = np.stack((arrA, arrA))
print(control_stack)
```

```python
# hstack()
hstck = np.hstack((arrA, arrA))
print(hstck)
# what this function is doing, is that it takes 2 of arrA and stack them horizontally.
```

```python
# vstack()
vstck = np.vstack((arrA,arrA))
print(vstck)
# previously, the two arrA arrays are stacked side by side, this time
# it is stacked vertically. Which means ontop of eachother.
```

```python
# dstack()
dstck = np.dstack((arrA, arrA))
print(dstck)
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)
# It printed two arrays. Both still has two dimensions
```

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)
# It has split into 3 arrays. All still having two dimensions, but the split was uneven
```

## Searching and replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

## Sorting and Filtering Arrays

```python
arrB = np.array([(7,6), (13,5), (7,3)])
arrB = np.sort(arrB)
print(arrB)
```

```python
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
print(arrB)
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```
