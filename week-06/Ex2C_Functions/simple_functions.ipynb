Working with defining custom functions.

```python
# Create function named favorite_things w/ 3 parameters
def favorite_things(name, movie, music):
    return f"Name: {name}\nMovie: {movie}\nMusic: {music}"
```

```python
info = favorite_things('Kevin', 'Grown Ups', 'Moonlight')
print(info)
```

```python
# create a function named why_im_here() w/ 1 parameter
def why_im_here(purpose):
    return purpose
```

```python
print(why_im_here("I joined Year Up United Data Analyst program because I wanted to start my career in it"))
```

```python
# create a function named favorite_places() w/ place and why
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}"
```

```python
print(favorite_place("Yosemite", "it has so much history to it and the hikes there are fantastic to go with family or friends"))
```
