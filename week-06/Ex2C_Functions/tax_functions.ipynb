Creating tax calculation function

## Social Security Tax

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

## Medicare Tax

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

## Federal Tax

```python
def get_federal_tax(gross_pay, code):
    tax_rate = 0
    if code == 0:
        tax_rate = .23
    elif code == 1:
        tax_rate = .21
    elif code == 2:
        tax_rate = .195
    elif code == 3:
        tax_rate = .185
    else:
        tax_rate = 0.18 - ((code - 4) * 0.005)
        
        return gross_pay * tax_rate
```

## Test function

```python
gross_pay = 750

print(f"Social Security Tax: ${get_soc_sec_tax(gross_pay):.2f}")
print(f"Medicare Tax: ${get_medicare_tax(gross_pay):.2f}")
print(f"Federal Tax: {get_federal_tax(gross_pay, 0)}")
```
