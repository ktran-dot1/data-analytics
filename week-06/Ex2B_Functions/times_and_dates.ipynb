```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
day_of_week = today.strftime("%A")
month = today.strftime("%m")
day = today.strftime("%d")
year = today.strftime("%y")
```

```python
print(f"Today is {day_of_week}, {month}/{day}/{year}")
```

# My birthday

```python
birthday = datetime.date(2004, 1, 1)
```

```python
print(birthday)
```

```python
month = birthday.strftime("%m")
day = birthday.strftime("%d")
year = birthday.strftime("%Y")
```

```python
print(f"My birthday is {month}/{day}/{year}")
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
month = ninety_d.strftime("%B")
day = ninety_d.strftime("%d")
year = ninety_d.strftime("%Y")
```

```python
print(f"90 days from today is {month} {day},{year}")
```

# Time for Dinner

```python
dinner_time = datetime.time(7, 30)
```

```python
hour = dinner_time.strftime("%I")
minute = dinner_time.strftime("%M")
```

```python
print(f"Let's meet for dinner at {hour}:{minute} PM")
```
