## Section 3.A Classes

### 3.A.1 What are classes?

- A **class** is like a blueprint used to create objects
    - Making an object from a class is called instantiation

```python
# Example:
class SimpleDog:
    '''Simple model for a dog'''

    def __init__(self, name, age, he_she):
        self.name = name
        self.age = age
        self.he_she = he_she

    def sit(self):
        print(f'{self.name} is now sitting')

my_dog = SimpleDog('Bobo', 3, 'He') # creating an instance
print(f"My dog's name is {my_dog.name}")
my_dog.sit()
```

## Section 3.B Class Attributes & Inheritance

### 3.B.5 Inheritance & child classes

- You can write child classes based on an existing parent class
    - Also referred to as subclass(child) and superclass(parent)
    - In the child's __init __() method, you'll call on the init method from the parent class 
    - Then set attributes specific to the child class
