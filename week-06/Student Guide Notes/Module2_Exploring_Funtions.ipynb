## Section 2.D Lambdas, Comprehensions & Generators

### 2.D.1 Lambda functions

- A **Lambda** is a small function used with a single expression
- The syntax for lambda function is:
    - lambda parameter, [parameter2, ...]: expression 

```python
# Example: using lambda to define a variable
x = lambda a, b: a * b # Put lambda func in () following with its arguments to use func immediately
print(x(5,6))
```

- map() and filter(), are designed to work with a function argument, and lambda is a concise way to supply it

- **map()** applies a function to every item and returns the results

```python
prices = [10, 25, 8, 40, 15]

discounted = list(map(lambda p: p * 0.8, prices)) # map() is applying it onto prices
print(discounted)
```

**filter()** keeps only items where the function returns True

```python
prices = [10, 25, 8, 40, 15]

expensive = list(filter(lambda p: p > 12, prices)) # filter() is applied to prices
print(expensive)
```

### 2.D.4 List comprehensions

- A list comprehension is a way to build a list from another collection. Achieving the same results as
map() and filter()
- It has three parts: an expression, a loop variable, and optional condition
    - expression **for** item in interable **if** condition

```python
# Example:
prices = [10, 25, 8, 40, 15]
# Transform (produces same result as map() example)
discounted = [p * 0.8 for p in prices]
print(discounted)
```

```python
# Example:
prices = [10, 25, 8, 40, 15]
# filter(produces the same result as filter())
expensive = [p for p in prices if p > 12]
print(expensive)
```

### 2.D.5 Generator expressions

- A **generator expression** produces a value one at a time rather than a complete list
- Similar to list comprehension, that uses [], but generator uses ()
- Generator expressions are helpful as arguments to aggregation funcs like sum(), max(), and min()


```python
# Examples
total = sum(p for p in prices) # output: 98
highest = max(p for p in prices) # output: 40
big_only = sum(p for p in prices if p > 12) # output: 80
```

- Generator expression also work with unpacking
- Use **list comprehension** when you need the results as a list. to index into it, loop over it, print, or pass it 
- Use **generator expression** when you are feeding the results direclty into aggregated functions

### 2.D.7 Generator Functions

- a **generator function** can be used to define a custom iterator that called when needed
    - An iterator is an object which executes over a sequence of items
- Uses **yield** keyword to define the return value for each iteration

```python
# Example:
def greeting():
    yield "Hello!"
    yield "How are you?"
    how_doing = (input("Your reply: ").lower())
    if how_doing == "good" or how_doing == "great":
        yield "That's good. What's new?"
    else:
        yield "Hey, how about this weather?"

for x in greeting():
    print(x)
```

```python
# 2nd example:
def four_count():
    x = 0
    while x > 4:
        yield x
        x += 1

for i in four_count():
    print(i)
```
