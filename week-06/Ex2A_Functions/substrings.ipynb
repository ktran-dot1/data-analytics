## Lab 1. Working with names - parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i+1:]
            print(f"Full name: {name}")
            print(f"First name: {first}")
            print(f"Last name: {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
parse_and_display(test_list[0])
```

```python
parse_and_display(test_list[1])
```

```python
parse_and_display(test_list[2])
```

```python
parse_and_display('Kevin Tran')
```

```python
def parse_and_display2(name):
    parts = name.split() # splits name with more than 1 word: 'Kevin Tran' --> ['Kevin', 'Tran']
    if len(parts) == 1: # single name
        print(f"First name: {parts[0]}")
    elif len(parts) == 2: # First and last name
        print(f"First name: {parts[0]}")
        print(f"Last name: {parts[1]}")
    elif len(parts) == 3: # first, middle, and last name
        print(f"First name: {parts[0]}")
        print(f"Middle name: {parts[1]}")
        print(f"Last name: {parts[2]}")
```

```python
pnd2_test = ['Lorde', 'Billie Eilish', 'Megan Thee Stallion']
```

```python
parse_and_display2(pnd2_test[0])
parse_and_display2(pnd2_test[1])
parse_and_display2(pnd2_test[2])
```
