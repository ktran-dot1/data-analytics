# Continuing with Classes

```python
class SimpleDog: 
    '''Simple model for a dog''' 

    def __init__(self, name, age, he_she):
        self.name = name
        self.age = age
        self.he_she = he_she
        self.behavior = 'good'
        
    def sit(self): # this is a method
        return f"{self.name} is now sitting"

    def describe_behavior(self):
        return f'{self.he_she} is a {self.behavior} dog'

    def birthday(self):
        if str(self.age).lower() == 'less than 1':
            self.age = 1
        else:
            self.age += 1
        print(f'{self.name} is now {self.age}')
        return f'Happy birthday, {self.name}'
```

```python
bs_dog = SimpleDog('Carrot', 8, 'He')
js_dog = SimpleDog('Lola', 'less than 1', 'She')
ds_dog = SimpleDog('Cookie', 7, 'She')
```

```python
class GoldenRetriever(SimpleDog):
    '''
    Defines a golden retriever, which characteristically has
    certain coloring and may be long-hair or extra fluffy.
    Many golden retriever enjoy swimming
    '''
    def __init__(self, name, age, he_she, coat='golden', swims='yes'):
        super().__init__(name, age, he_she)
        self.coat = coat
        self.swims = swims
    
```

```python
new_dog = GoldenRetriever('Chester', 3, 'He', coat="ruddy")
```

```python
print(new_dog.sit())
print(new_dog.describe_behavior())
print(new_dog.birthday())
```

```python
# new example: video game character

class GameCharacter: # parent class
    '''Basic model for all game characters'''
    def __init__(self, type, location):
        self.type = type
        self.location = location
    def char_profile(self):
        return f'''Character type: {self.type} 
        Location: {self.location}'''
```

```python
troll = GameCharacter('troll', 'Far West')
print(troll.char_profile())
```

```python
class Enemy(GameCharacter):
    '''Enemy troll, very nasty'''

    def __init__(self, type, location, strength):
        super().__init__(type, location)
        self.strength = strength
        self.defense = 70

    def alert(self):
        return f"Alert! You have encountered an enemy {self.type}"
```

```python
badtroll = Enemy("Evil Troll", "Far West", 75)
```

```python
print(badtroll.char_profile())
```

```python
# print(troll.alert()) # returns an error. no alert method with parent class
print(badtroll.alert())
```

```python
with open("sample.txt") as f: # consider best practice to use a with block to ensure files are auto closed
    print(f.read(100))
    print(f.read(100))
    print(f.read(100))
```

```python
with open("sample.txt") as f:
    print(f.readline())
    print(f.readline())
    print(f.readline())
```

```python
with open('sample.txt') as f:
    # for i in f:                # just looping directly creates odd results
    for i in range(10):          # effect is to read one line at a time
        print(f.readline())
```

```python
not_vanilla = []
vanilla = []
with open('sample.txt') as f:
    for i in f:
        if 'vanilla' not in i.lower():
            not_vanilla.append(tuple(i[0:-1].split(",")))
        else:
            vanilla.append(tuple(i[0:-1].split(",")))

    
print(not_vanilla)
print(vanilla)
```
