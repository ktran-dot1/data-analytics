## Defining custom functions (continued)

```python
def rsvp(name, phone, going, *attendees):
    response = [name, phone, going, attendees]
    return response
```

```python
rsvp("Marian", "555-123-4567", "Y", "Robin", "Marian")
```

```python
rsvp(
    input("Name of respondent: "),
    input("Phone number: "),
    input("Attending (Y/N): "),
    input("Name of attendees, including yourself: ") # this will always return one string
)
```

### arbitrary keyword argument using **

```python
def relationship_tracker(relationship, name, **kwargs):
    kwargs['relationship'] = relationship
    kwargs['name'] = name
    return kwargs
```

```python
person1 = relationship_tracker('cousin', 'Bea', nickname='gooseface')
print(person1)
```

```python
person2 = relationship_tracker('friend', 'Bilbo', height='short', feet='hairy')
print(person2)
```

## Working with anonymous functions (lambdas)

```python
(lambda a, b: a * b)(10, 5)
```

```python
# using a lambda in a variable definition

x = lambda a, b: a * b
print(x(10, 5))
```

```python
t_k = {
    'Ftok': lambda deg_f: 273.15 + (deg_f - 32) * 5 /9,
    'Ctok': lambda deg_c: 273.15 + (deg_c)
}

print(t_k['Ftok'](32))
```

```python
# using a Lambda inside a named function

def discount_calculator(pct_off):
    return lambda price: price * (1 - pct_off)
```

```python
pct_off20 = discount_calculator(.2) # 20% off

pct_off30 = discount_calculator(.3)

price = float(input('Enter price: '))

print(f"20% off: ${pct_off20(price):.2f}, members get 30% off: ${pct_off30(price):.2f}")
```

#### Using lambdas to supply a function as an argument

```python
prices = [10, 25, 8, 40, 15]

discounted = list(map(lambda p: p * 0.8, prices)) # using the map function to apply an expression across all prices

print(map(lambda p: p * 0.8, prices))
print(discounted)
```

```python
expensive = list(filter(lambda p: p > 12, prices)) # filter to only the items that match the condition
print(expensive) # [25, 40, 15]
```

### using list comprehensions

```python
discounted = [p * 0.8 for p in prices]
print(discounted)
```

```python
expensive = [p for p in prices if p > 12]
print(expensive)
```

### working with generator functions (using yield)

```python
def greeting():
    yield "Hello"
    yield "How are you?"
    how_doing = (input("Your reply: ").lower())
    if how_doing == "good" or how_doing == "great":
        yield "That's good. What's new?"
    else:
        yield "Hey, how about this weather?"
```

```python
for i in greeting():
    print(i)
```

```python
def jellybean():
    yield "This is a jellybean counter"
    jellybean = 0
    while jellybean <= 4:
        yield f' we have {jellybean} jellybeans'
        jellybean += 1
    yield "That's enough jellybeans"
```

```python
for j in jellybean():
    print(j)
```

### adding class methods

```python
class SimpleDog: 
    '''Simple model for a dog''' 

    def __init__(self, name, age, he_she):
        self.name = name
        self.age = age
        self.he_she = he_she
        self.behavior = 'good'
        
    def sit(self): # this is a method
        return f"{self.name} is now sitting"

    def describe_behavior(self):
        return f'{self.he_she} is a {self.behavior} dog'

    def birthday(self):
        if str(self.age).lower() == 'less than 1':
            self.age = 1
        else:
            self.age += 1
        print(f'{self.name} is now {self.age}')
        return f'Happy birthday, {self.name}'
```

```python
bs_dog = SimpleDog('Carrot', 8, 'He')
js_dog = SimpleDog('Lola', 'less than 1', 'She')
ds_dog = SimpleDog('Cookie', 7, 'She')
```

```python
print(bs_dog.birthday())
print(js_dog.birthday())
print(ds_dog.birthday())
```
