### Using random library

```python
import random
```

```python
a = random.randint(1, 100)
a
```

```python
print(a)
```

```python
b = []

for i in range(10):
    b.append(random.randint(1, 100)) # .append to get the 10 rand nums into the list

print(b)
```

```python
random.choice(b) # random element from the list
```

```python
random.choices(b, weights=[1,2,1,2,1,2,1,2,6,7], k=5) # weights: optional parameter to increase probability of retrieving that value
```

```python
print(b)
```

```python
random.shuffle(b)
print(b)
```

```python
random.sample(b, k=5) # takes the amount of value from the list and creates a new one
```

## Custom Functions

### working with def keywork

```python
# very simple example of a function:

def pizza():
    return "Yay pizza!"
```

```python
pizza()
```

```python
def pizza_v_burritos(food_name):
    if food_name == "pizza":
        msg = "Yay pizza!"
    elif food_name == "burritos":
        msg = "Yay burritos!"
    else:
        msg = "boooooo"
    return msg
```

```python
pizza_v_burritos('burritos')
```

```python
pizza_v_burritos('pizza')
```

```python
pizza_v_burritos('hotdog')
```

```python
# example of a function with 2 parameters
def place_order(item, qty):
    return f'you ordered {qty} of item {item}'
```

```python
place_order('Pizza', 2)
```

```python
place_order('hockey puck', 67)
```

```python
place_order(input("What would like to order? (pizza or burrito)"), input('How many?'))
```

## using keyword arguments

```python
# def place_order(item, qty):
#     return f'you ordered {qty} of item {item}'
```

```python
place_order(qty="7", item="burritos")
```

```python
def tax_calc(rate, amount, exempt='N'):
    if exempt != 'N':
        tax_amt = 0
    else:
        tax_amt = rate * amount
    return tax_amt
```

```python
tax_calc(.08, 122.45)
```

```python
tax_calc(.075, 58, 'exempt')
```

### working with an arbitrary number of arguments

```python
def greeting_kids(*arg):
    return arg
```

```python
greet = greeting_kids('Henry', 'Alice')

greet
```

```python
def send_greeting(whomst):
    for i in whomst:
        print(f'hi, {i}!')
```

```python
send_greeting(greet)
```
