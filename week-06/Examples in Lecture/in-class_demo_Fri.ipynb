# Error Handling

### Working with try / except

```python
# without try / except

first_num = float(input("% to calculate as decimal (provide a whole number): "))

pct_as_dec = first_num / 100 

print(pct_as_dec)
```

```python
# with try / except

try:
    fnum = float(input("Number to divide: "))
    snum = float(input("Second number to use as divisor: "))
    div_result = fnum / snum
except (ValueError, ZeroDivisionError): # can combine excepts or have them in one except
    print('Error: Must be a number other than zero')
except:
    print('Encountered an erro')
else:
    print(div_result)
finally:
    print('Run again to calculate another decimal value.')
```

### can nest try / except block

```python
try:
    f = open('sample_xyz.txt') # if file not found, in 'r' mode will throw an error
    try:
        sample_content = f.read()
    except:
        print('Cannot read from file')
    else:
        print(sample_content)
    finally:
        f.close()
except:
    print('Error, cannot retrieve file.')
```

```python
try:
    f = open('sample.txt') # still opening in 'r' mode
    try:
        sample_content = f.write('\n') # trying to write to the file, when its on 'r' mode
        sample_content = f.write("Customer survey was sent out via email mailing list")
    except:
        print('Cannot read from file')
    else:
        print(sample_content)
    finally:
        f.close()
except:
    print('Error, cannot retrieve file.')
```

### using try / except with a loop

```python
print('Division calculator:')
first_n = None

while first_n == None:
    try:
        first_n = float(input("First number to divide: "))
    except:
        print("Invalid input! Must be a number")

second_n = None
answer = None

while second_n == None:
    try:
        second_n = float(input("Second number to divide by: "))
        answer = first_n / second_n
    except ValueError:
        print('Must be a number.')
    except ZeroDivisionError:
        print('Number cannot be zero.')
        second_n = None # Set to None again to continue the loop, since second_n will hold '0'

print(f'{first_n} divide by {second_n} is equal to {answer:.2f}')
```

```python
# another example of try / except combined with a loop that's operating on file contents

try:
    with open("sample.txt") as f: # if file is not found then trigger except block
        contents = f.readlines() # reads line by line, putting them in a list
except FileNotFoundError:
    print("Error: file not found")
else:
    with open("sales_data.csv", "w") as output:
        for i in contents:
            if "," in i:
                output.write(i)
    print("Output complete.")
finally:
    print("Run program again to process another file.")
```
