## Array Basics

```python
fruits = np.array(['apple', 'banana', 'cherry', 'date'])
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(type(arr_1))
print(arr_1.dtype)
```

The letter 'U' mean its a Unicode string array, hold up to six characters.

```python
arr_2 = np.array(((10,11,12),(13,14,15)))
print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)
```

.T doesn't permanently changes the array. As you can see after recalling arr_2 it still stays the same

```python
arr_3 = np.array((
    (1,3,5),
    (2,4,6)
))
print(arr_3)
```

```python
arr_4 = np.array((
    arr_3[0,2],
    arr_3[1,1]
))
print(arr_4)
```

```python
arr_5 = np.ones((2,2,2))

print(arr_5)
```

arr_5: It is three dimensions because of the three brackets. Each dimension contains 2 values using '1'
